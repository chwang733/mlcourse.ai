## Observation: 
1. customers without international plan have higher retention rate, 
2. customers with less than 3 customer service call have higher retention rate

```python
#feature engineering
df['Churn predictor']=(df['International plan']=='True') & (df['Customer service calls'] > 3)
```

```python
pd.crosstab(df['Churn'],df['Churn predictor'], margins=True)
```
